x-coordinate is defined as increasing in the rightward direction, and the y-coordinate is defined as increasing in the downward direction. At each point in the image, the resulting gradient approximations can be combined to give the gradient magnitude, using:

```
def sobelOperators(raw):
    image = np.copy(raw)
    size = image.shape
    for i in range(1, size[0] - 1):
        for j in range(1, size[1] - 1):
            gx = (raw[i - 1][j - 1] + 2*raw[i][j - 1] + raw[i + 1][j - 1]) - (raw[i - 1][j + 1] + 2*raw[i][j + 1] + raw[i + 1][j + 1])
            gy = (raw[i - 1][j - 1] + 2*raw[i - 1][j] + raw[i - 1][j + 1]) - (raw[i + 1][j - 1] + 2*raw[i + 1][j] + raw[i + 1][j + 1])
            image[i][j] = min(255, np.sqrt(gx**2 + gy**2))
    return image
    pass
```

```
raw = cv2.cvtColor(cv2.imread("piano.jpg"), cv2.COLOR_BGR2GRAY)
raw = cv2.cvtColor(raw, cv2.COLOR_GRAY2RGB)
plt.imshow(raw)
plt.show()
```

```
%%time

image = cv2.cvtColor(cv2.imread("piano.jpg"), cv2.COLOR_BGR2GRAY)
image = sobelOperators(image)
image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
plt.imshow(image)
plt.show()
```
